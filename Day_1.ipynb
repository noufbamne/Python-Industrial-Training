```python
# number is +ve or -ve or zero
n = int(input("Enter number"))
if n < 0 :
  print(n," is -ve")
elif n > 0:
  print(n," is +ve")
else: #default when nothing else worked
  print(n,"is zero")
print("Day 1 ended")
```
