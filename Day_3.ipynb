```python
#print numbers greater than avg of series
s = pd.Series([5,12,6,7,9,22,35,65,9,11,23,67,89])
```

```python
avg = sum(s)/len(s)
```

```python
print("Average is: ",avg)
```

```python
for item in s:
  if item>=avg:
    print(item)
```
