#continue

```python
for i in range(1,6):
  if i==3:
    continue
  print(i)
print("Life goes on")
```

#break

```python
for i in range(1,6):
  if i == 3:
    break
  print(i)
print("Life goes on")
```

```python
for i in range(1,6):
  if i == 3:
    break
  print(i)
else:
  print("Life goes on")
```

1. Tea

0. Exit

```python
#Menu Driven Code
while True: #never stops on its own
  n = int(input("\n 1.Tea \n 2.Coffee \n 3.Water \n 0.Exit \n:")) #menu
  if n==1:
    print("Tea selected.")
  elif n==2:
    print("Coffee selected.")
  elif n==3:
    print("Water selected.")
  elif n==0:
    print("Exit......")
    break #take you out of loop
  else:
    print("Wrong choice .... ")
```
